```python
def bagi_list(input_list, input_bagi):
  """
  Fungsi untuk membagi list 1D menjadi 2D dengan ukuran chunk yang ditentukan.

  Args:
    input_list: List 1D yang berisi elemen-elemen yang ingin dibagi.
    input_bagi: Integer yang menentukan ukuran chunk untuk pembagian list.

  Returns:
    List 2D yang berisi list-list 1D dengan ukuran chunk yang ditentukan.
  """
  list_2d = []
  for i in range(0, len(input_list), input_bagi):
    list_2d.append(input_list[i:i+input_bagi])
  return list_2d

# Contoh penggunaan
input_list = [1, 2, 3, 4]
input_bagi = 2
output_list = bagi_list(input_list, input_bagi)
print(f"List 2D: {output_list}")
```

```python
all_col_tanggal_7 = bagi_list(all_col_tanggal, 7)
```

```python
from itertools import chain

def flatten_list(input_list):
  """
  Fungsi untuk meratakan list bersarang menggunakan `itertools.chain.from_iterable`.

  Args:
    input_list: List bersarang yang ingin diratakan.

  Returns:
    List 1D yang berisi elemen-elemen dari list bersarang.
  """
  return list(chain.from_iterable(input_list))

# Contoh penggunaan
input_list = [[1, 2], [3, 4], [5, 6]]
flattened_list = flatten_list(input_list)
print(f"List 1D: {flattened_list}")
```

```python
all_tanggal_7 = []
for j, col in enumerate(all_col_tanggal_7):
    all_tanggal_7.append([f"Minggu {j+1}-"+str(i*10) for i in range(1, 8)])

all_tanggal_7 = flatten_list(all_tanggal_7)
```

```python
all_beban = []
for col in all_col_tanggal:
    all_beban.append(df[col].tolist())

all_beban = flatten_list(all_beban)
all_beban_min = min(all_beban)
all_beban_max = max(all_beban)
```

```python
def normalize(x, xmin, xmax):
    return (x-xmin)/(xmax-xmin)
```

```python
for col in all_col_tanggal:
    df[col] = [normalize(df[col][i], all_beban_min, all_beban_max) for i in range(len(df))]
```
