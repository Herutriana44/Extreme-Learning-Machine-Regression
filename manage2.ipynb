```python
import pandas as pd

df = pd.read_excel('DATASET MANAGE 1.xlsx')
```

```python
df.head()
```

```python
all_col_tanggal = df.columns[1:]
```

```python
all_col_tanggal
```

```python
all_df_temp = []
for col_ind, col in enumerate(all_col_tanggal):
    col_tanggal_ind = col_ind
    df_temp = pd.DataFrame({
        'Jam' : df['Jam'].tolist(),
        'Hari' : all_col_tanggal[col_tanggal_ind],
        'Beban' : df[all_col_tanggal[col_tanggal_ind]].tolist()
    })

    all_df_temp.append(df_temp)

df_2 = pd.concat(all_df_temp)
```

```python
df_2
```

```python
df_2 = df_2.reset_index(drop=True)
```

```python
df_2['Minggu'] = [int(str(df_2['Hari'][i]).split('-')[0].replace('Minggu ', '')) for i in range(len(df_2))]
```

```python
df_2['Hari'] = [int(str(df_2['Hari'][i]).split('-')[1]) for i in range(len(df_2))]
```

```python
df_2
```

```python
df_2.to_excel('DATASET MANAGE 2.xlsx', index=False)
```
